## BAR CHART

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
company = ['Google','Amazon','MicroSoft','Facebook']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

### bar chart don't render string values, so we create an array to represent x - axis

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
# here bar don't contain company name, we've to use xtick function for that
```

```python
plt.xticks(xpos,company)   # this will replace the 0,1,2,3 values with company list
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label='Revenue')  # label is to 
plt.legend()
```

### To plot profit in same chat - side by side bar chart

```python
plt.xticks(xpos, company)   # for x axis ticks
plt.ylabel('Profit and Revenue (bln)')
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label = 'Revenue')  # first bar graph
plt.bar(xpos, profit, label = 'Profit')   # second bar graph
plt.legend()
# Here we are getting a stacked bar chart, because we have provided xpos same for both the graphs.
# To make these graphs side by side bars, we have to pass different xpos
```

### To get sidebyside bar chart, we have to control the xpos

```python
plt.xticks(xpos, company)
plt.ylabel('Profit and Revenue (bln)')
plt.title('US Tech company stocks')
plt.bar(xpos-0.2, revenue, label = 'Revenue')
plt.bar(xpos+0.2, profit, label= 'Profit')

# Can't we increase the graph size? 
```

### To make it a perfect chart (Use width property)

```python
plt.xticks(xpos, company)
plt.ylabel('Profit and Revenue (bln)')
plt.title('US Tech company stocks')
plt.bar(xpos-0.2, revenue, label = 'Revenue', width = 0.4)
plt.bar(xpos+0.2, profit, label= 'Profit', width = 0.4)
```

### Horizontal Bar Chart (Use of barh)

```python
plt.yticks(xpos, company)
plt.barh(xpos-0.2, revenue, label= 'Revenue')
plt.barh(xpos+0.2, profit, label = 'Profit')
plt.legend()
```
